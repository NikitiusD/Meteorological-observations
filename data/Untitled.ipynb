```python
import pandas as pd
from statistics import mode
```

```python
cities = ['Atlanta', 'Boston', 'Chicago', 'Detroit', 'Los Angeles', 'Miami', 'New York', 'Seattle']
columns = ['datetime'] + cities
```

```python
city_attributes = pd.read_csv('city_attributes.csv')
city_attributes = city_attributes[city_attributes['City'].isin(cities)]
city_attributes.reset_index(drop=True, inplace=True)
# city_attributes.drop(['Country'], axis=1, inplace=True)

city_attributes.to_csv('city_attributes_.csv')
print(city_attributes.shape)
city_attributes
```

```python
temperature = pd.read_csv('temperature.csv', parse_dates=['datetime'])
temperature = temperature[columns]
temperature['date'] = temperature['datetime'].dt.date
temperature = temperature[temperature['date'] >= pd.to_datetime('2015-01-01').date()]
temperature = temperature.groupby('date', as_index=False)[temperature.columns[1:-1]].mean()

temperature.to_csv('temperature_.csv')
print(temperature.shape)
temperature.head()
```
